Дан массив (10,2) координат, найти расстояние от каждой точки до каждой

```python
import scipy.spatial

Z = np.random.random((10,2))
D = scipy.spatial.distance.cdist(Z,Z)
print(D)
```

Случайно расположить p элементов в 2D массив

```python
n = 5
p = 3
X = np.zeros((n,n))
np.put(X, np.random.choice(range(n*n), p, replace=False), 1)
X
```

Дан вектор [1, 2, 3, 4, 5], построить новый вектор с тремя нулями между каждым значением

```python
x = np.array([1,2,3,4,5])
cnt = 3
y = np.zeros(len(x) + (len(x)-1)*(cnt))
y[::cnt+1] = x
y
```

Найти наиболее частое значение в массиве

```python
x = np.random.randint(0,5,10)
print(np.bincount(x).argmax())
```

Найти n наибольших значений в массиве

```python
x = np.arange(10000)
np.random.shuffle(x)
n = 5

x[np.argpartition(-x,n)[:n]]
```

## Pandas - фильтрация данных

Считать данные по  [адресу](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 
Присвоить переменной chipo

```python
import pandas as pd
chipo = pd.read_csv('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv', sep='\t')
chipo.head()
```

```python
chipo.info()
```

Как много продуктов стоят более чем $10.00?

```python
chipo['item_price'] = chipo['item_price'].replace('[\$,]', '', regex=True).astype(float)
len(chipo[chipo['item_price'] > 10.00])
```

Какова цена каждого предмета? Напечатать столбцы item_name и item_price

```python
for index, row in chipo[['item_name','item_price']].iterrows():
    print("The name of the item is {} prece is {}".format(row['item_name'], row['item_price']))
```

Отсортировать по имени предмета

```python
chipo.sort_values(by='item_name').head()
```

Сколько было заказано самого дорогого предмета?

```python
chipo.iloc[chipo['item_price'].idxmax()]['quantity']
```

Сколько раз был заказан предмет "Veggie Salad Bowl"?

```python
chipo[chipo['item_name'] == 'Veggie Salad Bowl']['quantity'].sum()
```

Сколько раз люди заказали более чем 1 предмет "Canned Soda"?

```python
len(chipo[(chipo['item_name'] == 'Canned Soda') & (chipo['quantity'] > 1)])
```
